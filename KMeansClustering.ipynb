# K-Means Clustering Model

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib import style
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

```python
x, y_true=make_blobs(n_samples=800,centers=5,cluster_std=0.45,random_state=0)
plt.scatter(x[:,0],x[:,1],s=30)
```

```python
# With 3 centres

style.use('ggplot')
func=KMeans(n_clusters=3)
func.fit(x)
y_func=func.predict(x)
plt.scatter(x[:,0],x[:,1],c=y_func,s=15)

#Centers generation and colouring 

centers=func.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='blue',s=75,alpha=1)
plt.show()
```

```python
# With 5 centres

style.use('dark_background')
func=KMeans(n_clusters=5)
func.fit(x)
y_func=func.predict(x)
plt.scatter(x[:,0],x[:,1],c=y_func,s=15)

#Centers generation and colouring 

centers=func.cluster_centers_
plt.scatter(centers[:,0],centers[:,1],c='white',s=75,alpha=1)
plt.show()
```
